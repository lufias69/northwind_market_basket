```python
import pandas as pd
```

# Load data dan Praproses

```python
df = pd.read_excel("data/products.xlsx")
product_dict = dict(zip(df.productID.tolist(),df.productName.tolist()))
# df
```

```python
df = pd.read_excel("data/order_details.xlsx")
orderID = df.orderID.tolist()
productID = df.productID.tolist()
productName = list()
code = ["x" for i in productID]

Quantity = df.quantity.tolist()
for i in productID:
    productName.append(product_dict[i])
# df

df = pd.DataFrame.from_dict({"orderID":orderID, "productName":productName, "Quantity":Quantity, "code":code})
df
```

# Membuat Basket

```python
basket = (df[df['code'] =="x"].groupby(['orderID', 'productName'])['Quantity']
          .sum().unstack().reset_index().fillna(0)
          .set_index('orderID'))
total_transaksi = len(basket)


# print(basket.head())
def encode_units(x):
    if x <= 0:
        return 0
    if x >= 1:
        return 1
basket_sets = basket.applymap(encode_units) #akhir
basket_sets
```

# Proses Apriori

```python
mui_sup = 0.003  #otak atik disini
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules


# Build up the frequent items 
frequent_itemsets = apriori(basket_sets, min_support=mui_sup, use_colnames=True)
frequent_itemsets
```

# Membuat rules

```python
# Create the rules
rules = association_rules(frequent_itemsets, metric="lift") #, min_threshold=0.5
rules =  rules.drop(['lift', 'leverage','conviction','antecedent support','consequent support'], axis=1)
rules
```
